# House Price Prediction:

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#To show all the rows of the dataset.
pd.pandas.set_option('display.max_columns',None)
```

```python
#Importing the dataset.
dataset=pd.read_csv('train.csv')
```

```python
#Displaying the top 5 rows
dataset.head()
```

### Missing Values:

```python
#Creating a list of features with null values
features_null=[features for features in dataset.columns if dataset[features].isnull().sum()>1]
features_null
```

```python
#Finding the percentage of null values present in each features.
for feature in features_null:
    print(feature, np.round(dataset[feature].isnull().mean(),4), "% missing values.")
```

## Trying to find the realtionship between Sale price and missing values:

Plotting some digrams to understand this better:

```python
for feature in features_null:
    data1=dataset.copy()
    
    #Lets replace the values of missing values with 1 and alternative with 0.
    data1[feature]=np.where(data1[feature].isnull(),1,0)
    
    #Calculating the median of the sale price where the information is missing or present.
    data1.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

As shown in the above diagrams the relation between missing values and dependent feature is very strong.  So, we have to replace these missing values with appropriate values which will done in feature engineering part.

### Numerical Variables:

```python
#Creating a list of numerical variables:
num_feature=[features for features in dataset.columns if dataset[features].dtypes!='O']

#Number of numerical features are:
print(f"There are {len(num_feature)} numerical features.")

#showing the numerical features alone as a dataset:
dataset[num_feature].head()
```

#### Handling the Temporal features:
Here there are 4 types of year features in this dataset, this changes every year so this can be considered as temporal features.

```python
# Temporal features related to year present our dataset can be initialised as:
year_temp_feature=[features for features in num_feature if 'Yr' in features or 'Year' in features ]
year_temp_feature
for feature in year_temp_feature:
    print(feature,dataset[feature].unique())
```

```python
#Trying to fetch information from YrSold feature:
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.title('House Price V/S Year of sold')
plt.xlabel('Year of sold')
plt.ylabel('House Price')
    
```

Here the house price is dicreasing over the year, which gives us a wrong insight. So, we will have to follow some other way to find a trend in the data. 

```python
# Trying to find the difference between YrSold and all the other year features,
# and plotting it against the Sale Price
for feature in year_temp_feature:
    if feature!='YrSold':
        data=dataset.copy()
        
        #changing all the feature values with the difference with the YrSold:
        data[feature]=data['YrSold']-data[feature]
        
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel("Year of Sold")
        plt.title(feature+" v/s Year of Sold")
        plt.show()
```

The above diagrams clearly show that with age of the house increasing the price also dicreases.

#### 1. Discrete Numerical Features:

```python
discrete_features=[features for features in num_feature if len(dataset[features].unique())<25 and features not in year_temp_feature+['Id']]
print("Number of Dicrete feaures: ",len(discrete_features))
```

```python
#Plotting the discrete features against the meadian of SalePrice:
for feature in discrete_features:
    dataset.groupby(feature)["SalePrice"].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature+" v/s SalePrice")
    plt.show()
```

#### 2.Continuous features:

```python
#Listing the continuous features in the numerical features.
continuous_features=[features for features in num_feature if features not in discrete_features+year_temp_feature+['Id']]
print(f"There are {len(continuous_features)} continuous features.")
continuous_features
```

```python
for feature in continuous_features:
    data=dataset.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.title(feature)
    plt.show()
```

Here not all the features follow Gaussian distribution. Some of them are skewed data as well. Now, converting these distributions into a Gaussian or Normal distribution will be helpful in linear model prediction.

```python
#Applying log normal distribution for all continuous features
for feature in continuous_features:
    data = dataset.copy()
    
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data['SalePrice']=np.log(data['SalePrice'])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.title(feature+" v/s SalePrice")
        plt.show()
```

### Outliers:

```python
for feature in continuous_features:
    data=dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.xlabel(feature)
        plt.title(feature)
        plt.show()
```
