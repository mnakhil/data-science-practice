# Feature Selection - House price prediction:
Here we basically identifies the relevent features for model building.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

#libraries required for feature selection:
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

#For showing all the columns present in the dataset
pd.pandas.set_option('display.max_columns',None)
```

```python
#importing data for feature selection
dataset=pd.read_csv('x_train.csv')
```

```python
dataset.head()
```

```python
# Choosing the dependent feature
y_train=dataset['SalePrice']
```

```python
#Dropping Id and SalePrice in x_train
x_train=dataset.drop(['Id','SalePrice','Unnamed: 0'],axis=1)
x_train.head()
```

```python
# Lasso Regression model is used here.
# SelectFromModel will help us to select the relevent features from the model.

feature_sel_model=SelectFromModel(Lasso(alpha=0.005,random_state=0)) #remember to set the seed
feature_sel_model.fit(x_train,y_train)
```

```python
feature_sel_model.get_support()
```

```python
#Storing the selected feature in a list

selected_features=x_train.columns[(feature_sel_model.get_support())]
selected_features
```

```python
# Displaying some of the details:
print("Total number of features in the train data:",len(x_train.columns))
print("Number of selected features: ",len(selected_features))
print("Number of features ignored: ",len(x_train.columns)-len(selected_features))
```

```python
x_train=x_train[selected_features]
x_train.head()
```

```python
# Storing the data inside a csv file
x_train.to_csv('fs_train.csv')
```
