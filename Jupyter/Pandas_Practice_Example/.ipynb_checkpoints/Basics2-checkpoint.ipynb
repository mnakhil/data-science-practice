## Basics of pandas using an imported CSV file:
A couple of csv files are downloaded from stackoverflow and using them basic operations of pandas are done here.

```python
import pandas as pd
df=pd.read_csv('survey_results_public.csv')
df.head()
```

```python
df.columns
```

##### Finding the count of elements found inside a column: 

```python
#The below command gives us the total count of each element present in df
df['Gender'].value_counts()
```

#### loc() and iloc(): 

```python
#using iloc to display first two rows
df.iloc[[0,1]]
```

```python
#displaying the rows as well as the specific columns that we wanted
df.iloc[[0,2],[3,6]]
```

```python
#displaying the desired rows using loc
df.loc[[0,2]]
```

```python
#To display specific rows and columns using loc
df.loc[[0,1,2],['Country','EdLevel']]
```

```python
#reading schema csv file using pandas
df_schema=pd.read_csv('survey_results_schema.csv')
df_schema.head()
```

```python
# Find out the size of the data frame using .shape function
df_schema.shape
```

```python
#Find all the atributes of first row
df.loc[0 ]
```

```python
#Find out an atribute value of a particular row or some different rows
df.loc[[0],'Gender']
```

```python
df.loc[[0,1,2],'Gender'] #Finds gender of 0,1 and 2 columns
```

```python
df.loc[0:2,'Age']
```

```python
#we can use indexing between two columns using loc as shown below
df.loc[0:2,'Country':'EdLevel']
```
