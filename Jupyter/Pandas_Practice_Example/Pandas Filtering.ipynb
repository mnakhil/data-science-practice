## Pandas Filtering:
This is used to filter out certain rows and columns that follows a certain condition.


```python
import pandas as pd
people={'first':['Akhil','Anusree','Ancy'],
       'second':['Muraleedharan','Muraleedharan','S Kumar'],
       'email':['akhil@gmail.com','anu@gmail.com','ancy@gmail.com']}
df=pd.DataFrame(people)
df
```

```python
#filtering done using normal indexing method
filt=(df['second']=='Muraleedharan')
df[filt]
```

```python
#filteing done using loc function
df.loc[filt,'email']
```

```python
#operator 'and' can't be used in pandas '&' is used instead
filt1=(df['first']=='Akhil') & (df['second']=='Muraleedharan' )
df.loc[filt1,'email']
```

```python
filt1=(df['first']=='Akhil') | (df['second']=='Muraleedharan' )
df.loc[filt1,'email']
```

```python
df1=pd.read_csv('survey_results_public.csv')
df1.head()
```

```python
#filtering out people with a salary of more than 80000
hisal=(df1['ConvertedCompYearly'] > 80000)
df1.loc[hisal]
```

```python
df1['Country'].value_counts()
```

```python
# New list is made with names of certain countries
countries=['Dominica','Papua New Guinea', 'Tuvalu','Brunei Darussalam']
filt2=(df1['Country'].isin(countries))# isin method is used to check whether the elements in the given list is present in the dataframe or not.
df1.loc[filt2,['MainBranch','Country']]
```

```python
# Find out the elements that contains a particular string in it.
master=df1['EdLevel'].str.contains('M.A.',na=False)
df1.loc[master]
```
